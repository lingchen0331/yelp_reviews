### Business Aggregated Data

Firstly, we want to select top 5 states in United States which has the largest number of reviews, and we can start reviewing the data from these states.

```python
business_location = []

for line in gzip.open("data/business.json.txt.gz", 'rt', encoding='utf-8'):
    business_info = json.loads(line.strip())
    if business_info["state"] in list_states:
        business_location.append(business_info["state"])
```

```python
print(business_location[:10])
```

```python
counter_states = Counter(business_location)
counter_states_most_common = counter_states.most_common(5)

labels_total_posts = []; numbers_total_posts = []
for i in range(5):
    labels_total_posts.append(counter_states_most_common[i][0])
    numbers_total_posts.append(counter_states_most_common[i][1])
```

```python
plt.bar(np.arange(5), numbers_total_posts, 0.55, alpha=0.4, color='b')
plt.xticks(range(5),labels_total_posts)
plt.title("The Number of Total Business in Each of Selected State")
plt.xlabel("States")
plt.ylabel("The number of Business")
plt.savefig('stats/Figure1.png')
plt.show()
```

Then, I will select reviews from these 5 states (AZ, NV, NC, OH and PA).

```python
# Create two lists for storing the name of the selected states and 
# the number of business of of the selected states.
selected_states = []
selected_states_business_number = []
for i in range(5):
    selected_states.append(counter_states.most_common(5)[i][0])
    selected_states_business_number.append(counter_states.most_common(5)[i][1])
    
# zip the two lists and create a new dict
selected_states_bId = dict(zip(selected_states, selected_states_business_number))
```

```python
business_id_states = []
business_rates_states = []

for i in range(len(selected_states)):
    business_id_states.append([])
    business_rates_states.append([])
```

```python
for line in gzip.open("data/business.json.txt.gz", 'rt', encoding='utf-8'):
    business_info = json.loads(line.strip())
    for k in range(len(selected_states)):
        if business_info["state"] == selected_states[k]:
            business_id_states[k].append(business_info["business_id"])
            business_rates_states[k].append(business_info["stars"])
```

Now, we have two lists representing the selected business ID and the corresponding rates(stars), respectively. 

```python
colors = ['red', 'blue', 'yellow', 'green', 'purple']
data_hist = []
for i in range(len(business_rates_states)):
    data_hist.append(np.random.normal(business_rates_states[i]))

fig, ax = plt.subplots()
bins = np.linspace(1, 5, 100)

for j in range(len(business_rates_states)):
    ax.hist(data_hist[j], bins=bins, color=colors[j], label=selected_states[j],
        alpha=0.5, histtype='stepfilled')

ax.legend(loc='upper left')
ax.margins(0)
plt.xlabel("Rates (from 1 to 5)")
plt.ylabel("Number of Business")
plt.title('Histogram for number of rates distribution in selected states')
plt.savefig('stats/Figure2.png')
plt.show()
```

```python
means = []; std = [];mins = []; maxes = []
for k in range(len(business_rates_states)):
    mins.append(np.min(business_rates_states[k]))
    maxes.append(np.max(business_rates_states[k]))
    means.append(np.mean(business_rates_states[k]))
    std.append(np.std(business_rates_states[k]))

mean_min = []; max_mean = [];
for a in range(5):
    mean_min.append(means[a] - mins[a])
    max_mean.append(maxes[a] - means[a])
    
fig, ax1 = plt.subplots(1,1)

x = [1,2,3,4,5]
plt.errorbar(np.arange(5), means, std, fmt='ok', lw=5)
plt.errorbar(np.arange(5), means, [mean_min, max_mean], fmt='.k', ecolor='gray', lw=1)
plt.xlim(-1, 5)
states = ['', 'AZ', 'NV', 'NC', 'OH', 'PA']
ax1.set_xticklabels(states, minor = False)
plt.ylabel("Average rate of restaurants")
plt.title('Error Bar Plot for mean, std, maximum and minimum')
plt.gcf().set_size_inches(7.0, 5.0)
plt.savefig('stats/Figure3.png')
plt.show()
```

From the two plots, we can clearly see that the average rate of business is between 3.5 and 4.0, and approximately around 3.6, which shows that the average is a little bit above the arithmetic average, and the range between maximum - average and average - minimum is around the same. 

Next, let's analyze people's preference at each selected state. Specifically, I will pick top 500 business in the five selected state, and try to find the most favorable business type. 

```python
rating_list = []
for i in range(5):
    rating_list.append(dict(zip(business_id_states[i], business_rates_states[i])))
```

```python
# Sort the rating
sorted_rating_list = []
for i in range(len(rating_list)):
    sorted_rating_list.append(sorted(rating_list[i].items(), key=lambda value: value[1], reverse = True))
```

```python
# Take first 500 business (top_rated)
for i in range(len(sorted_rating_list)):
    sorted_rating_list[i] = sorted_rating_list[i][:500]
```

```python
# For example, we can take a look at the business "categories", to check if there are any
# common grounds between high-rated business in each area
high_rating_business_id = []
for i in range(5):
    high_rating_business_id.append([])
    
for a in range(len(high_rating_business_id)):    
    for i in range(len(sorted_rating_list[0])):
        high_rating_business_id[a].append(sorted_rating_list[a][i][0])
```

Next, based on the business id, we can get the category of those business which has the high rate.

```python
high_rating_business_categories = []
for i in range(5):
    high_rating_business_categories.append([])

# Getting categories based on business ID
for line in gzip.open("data/business.json.txt.gz", 'rt', encoding='utf-8'):
    business_info = json.loads(line.strip())
    for a in range(len(high_rating_business_categories)):    
        for item in high_rating_business_id[a]:
            if business_info["business_id"] == item:
                high_rating_business_categories[a].append(business_info["categories"])
```

```python
# Then, we need to turn the nested list to a flat list
for it in range(5):
    high_rating_business_categories[it] = [i for j in high_rating_business_categories[it] for i in j]
```

```python
category_corpus = []
category_corpus_common = []
for i in range(5):
    category_corpus.append([])
    category_corpus_common.append([])

for j in range(5):    
    category_corpus[j] = Counter(high_rating_business_categories[j])
    category_corpus_common[j] = category_corpus[j].most_common(10)
```

```python
labels = []
numbers = []
for i in range(5):
    labels.append([])
    numbers.append([])
    
for k in range(5):    
    for j in range(10):
        labels[k].append(category_corpus_common[k][j][0])
        numbers[k].append(category_corpus_common[k][j][1])
```

```python
opacity = 0.6
colors = ['r', 'g', 'b', 'brown', 'orange']

for i in range(5):
    plt.subplot(5, 1, i+1)
    plt.barh(np.arange(10), numbers[i], 0.55, alpha=opacity, color=colors[i])
    plt.yticks(range(10),labels[i])
    plt.xlim([0,120])
    plt.title("The Most Popular Business Categories in " + selected_states[i])

plt.gcf().set_size_inches(9.0, 22.0)
plt.savefig('stats/Figure4.png')
plt.show()
```

Next, let's take a look at if different states have the same preference.

```python
labels_flat = [i for j in labels for i in j]
labels_frequency = Counter(labels_flat)
labels_frequency = labels_frequency.most_common()
```

```python
labels_frequency_plot = []; numbers_frequency_plot = []
for i in range(len(labels_frequency)):
    labels_frequency_plot.append(labels_frequency[i][0])
    numbers_frequency_plot.append(labels_frequency[i][1])

plt.barh(np.arange(len(labels_frequency_plot)), numbers_frequency_plot, 0.55, alpha=opacity, color = 'b')
plt.yticks(range(len(labels_frequency_plot)),labels_frequency_plot)
plt.title("The Most Popular Business Categories in Selected States")
plt.gcf().set_size_inches(9.0, 9.0)
plt.savefig('stats/Figure5.png')
plt.show()
```

From the above plots, we can see in the selected states, residents tend to give high rates in the following categories: Active Life, Automotive, Shopping, Health & Medical and Beauty & Spa among the five selected states. However, there are examples showing people from one state have preference at certain business; for example, only people who living in North Carolina generally give Pets and Fitness related business high score. 

Next, I am interested in the location of high-rated business in the selected area, and I will use Arizona as an example. Below is the Arizona wealth level by zip code. 

```python
lines = []
with open("data/zipcode_wealthlevel.txt", 'r', encoding="utf-16") as file:
    for l in file:
        lines.append(l.strip().split('\t'))
        
zipcode = []; cities = []; income = []
for i in range(len(lines)):
    zipcode.append(lines[i][1])
    cities.append(lines[i][3].split(",")[0].replace('"', ""))
    income.append(float(sub(r'[^\d.]', '', lines[i][5].replace('"', "").replace(' ', ""))))
```

```python
zip_income = dict(zip(zipcode, income))
```

```python
# I already generated the list of businesses with high average rating
# Based on the business id, I am able to get the corresponding zip code
city = []
postalcode = []
for line in gzip.open("data/business.json.txt.gz", 'rt', encoding='utf-8'):
    location_info = json.loads(line.strip())
    if location_info["business_id"] in high_rating_business_id[0]:
        city.append(location_info["city"])
        if location_info["postal_code"] != "":
            postalcode.append(location_info["postal_code"])
```

```python
high_rate_common_city_AZ = Counter(city)
high_rate_common_city_AZ = high_rate_common_city_AZ.most_common()

high_rate_common_zipcode_AZ = Counter(postalcode)
high_rate_common_zipcode_AZ = high_rate_common_zipcode_AZ.most_common()
```

```python
# For now, we have a sorted list of zip codes with high rating business
# We can create a new list to store these data
zipcode_sorted = []

for i in range(len(high_rate_common_zipcode_AZ)):
    zipcode_sorted.append(high_rate_common_zipcode_AZ[i][0])
```

### Actual Reviews of Business

Since we have already analyzed the characteristics of rates in selected states, the next step would be getting the reviews corresponding to the business id in the selected states. 

To do that, we need to introduce another dataset - review.json.

```python
actual_stars_AZ = []
reviews_AZ = []

for line in gzip.open("data/review.json.txt.gz", 'rt', encoding='utf-8'):
    review_info = json.loads(line.strip())
    if review_info["business_id"] in business_id_states[0]:
        actual_stars_AZ.append(review_info["stars"])
        reviews_AZ.append(review_info["text"])
```

```python
fig, ax = plt.subplots()
bins = np.linspace(0, 6, 20)

ax.hist(actual_stars_AZ, bins=bins, color='b', label='AZ', alpha=0.5, histtype='stepfilled')

ax.legend(loc='upper left')
ax.margins(0)
    
plt.xlabel("Rates (from 1 to 5)")
plt.ylabel("Number of Business")
plt.ylim([0, 240000])
plt.title('Histogram for number of actual stars distribution in selected states')
plt.savefig('stats/Figure7.png')
plt.show()
```

Then, I am interested in the text sentiment analysis for those reviews. Some users tend to only leave stars without comments, and some people's comments are really subjective. From the review itself, I want to compare the result of text sentiment analysis with the actual rating

```python
sentiment_analysis_polarity = []
sentiment_analysis_subjectivity = []

for i in range(len(reviews_AZ)):
    sentiment_analysis_polarity.append(TextBlob(reviews_AZ[i]).sentiment.polarity)
    sentiment_analysis_subjectivity.append(TextBlob(reviews_AZ[i]).sentiment.subjectivity)
```

```python
# Since I used [-1, 1] for the scale of polarity, I need to transfer this scale to [1, 5]
for i in range(len(sentiment_analysis_polarity)):
    sentiment_analysis_polarity[i] = sentiment_analysis_polarity[i]*2 + 3
```

```python
len(sentiment_analysis_polarity)
```

```python
fig, ax = plt.subplots()
bins = np.linspace(0, 5, 100)

ax.hist(sentiment_analysis_polarity, bins=bins, color='r', label='AZ',alpha=0.5, histtype='stepfilled')

ax.legend(loc='upper left')
ax.margins(0)
x = plt.axvline(np.mean(sentiment_analysis_polarity), color='black', linestyle='dashed', linewidth=2, label = 'Average Rating')
plt.xlabel("Rates (from 1 to 5)")
plt.ylabel("Number of Business")
plt.legend()
plt.title('Histogram for number of rates distribution in selected states')
plt.savefig('stats/Figure8.png')
plt.show()
```

Comparing with this plot with the Figure 2, we cannot see any significant difference
